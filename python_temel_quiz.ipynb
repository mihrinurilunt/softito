## Soru 1 — Değişkenler, Veri Tipleri ve String İşlemleri

Aşağıdaki bilgileri kullanarak bir **kişi kartı** oluşturun:

- `isim = "ahmet yılmaz"` → baş harfleri büyük yapın (`title()`)
- `dogum_yili = 1998` → 2025 yılına göre yaşı hesaplayın
- `boy = 1.75` (metre cinsinden)
- `aktif = True`

**Görev:** Aşağıdaki çıktıyı üretecek kodu yazın:

```
=== KİŞİ KARTI ===
İsim   : Ahmet Yılmaz
Yaş    : 27
Boy    : 1.75 m
Durum  : Aktif
```

> 💡 İpucu: f-string ve `title()` kullanın. Durum için ternary if ile `"Aktif"` / `"Pasif"` yazdırın.

````python
# KODUNUZU BURAYA YAZIN
kisi_karti = {
    "isim": "ahmet yılmaz",
    "dogum_yili": 1998,
    "boy": 1.75,
    "aktif": True
}
kisi_karti["isim"] = kisi_karti["isim"].capitalize()
kisi_karti["yas"] = 2025 - kisi_karti["dogum_yili"]
kisi_karti["durum"] = "Aktif" if kisi_karti["aktif"] else "Pasif"
print(kisi_karti)

````

## Soru 2 — Koşullar ve Operatörler

Bir market uygulaması için **indirim hesaplayıcı** yazın:

| Toplam Tutar | İndirim Oranı |
|---|---|
| 500 TL ve üzeri | %20 |
| 200 TL ve üzeri | %10 |
| 100 TL ve üzeri | %5 |
| 100 TL altı | İndirim yok |

**Görev:** `tutar` değişkenini alıp indirim sonrası ödenmesi gereken tutarı yazdıran kodu yazın.

```
# tutar = 350 için beklenen çıktı:
Tutar    : 350 TL
İndirim  : %10 (35.0 TL)
Ödenecek : 315.0 TL
```

> 💡 İpucu: `if / elif / else` kullanın.

````python
def indirim_hesaplayici(tutar):
  if tutar >= 500:
    oran = 20
  elif tutar >= 200:
    oran = 10
  elif tutar >= 100:
    oran = 5
  else:
    oran = 0

  indirim_miktari = tutar * oran / 100
  odenecek = tutar - indirim_miktari

  if oran > 0:
    print(f"Tutar    : {tutar} TL")
    print(f"İndirim  : %{oran} ({indirim_miktari} TL)")
    print(f"Ödenecek : {odenecek} TL")

tutar = 350
indirim_hesaplayici(tutar)
````

## Soru 3 — Döngüler ve Listeler

Aşağıdaki öğrenci notları listesi verilmiştir:

```python
notlar = [72, 45, 88, 91, 55, 63, 78, 40, 95, 60]
```

**Görev:** Tek bir döngü kullanarak şunları hesaplayın ve yazdırın:

1. Geçen öğrenci sayısı (50 ve üzeri)
2. Kalan öğrenci sayısı (50 altı)
3. En yüksek not
4. Notların ortalaması
5. 80 ve üzeri not alan öğrencilerin notlarından oluşan yeni liste

```
# Beklenen çıktı:
Geçen    : 8 öğrenci
Kalan    : 2 öğrenci
En yüksek: 95
Ortalama : 68.7
Yüksek notlar: [88, 91, 95]
```

> 💡 İpucu: `for` döngüsü + sayaç değişkenleri kullanın. Son madde için liste comprehension deneyin.

````python
notlar = [72, 45, 88, 91, 55, 63, 78, 40, 95, 60]

en_yuksek = 0
yuksek_notlar = []
toplam = 0
gecen = 0
kalan = 0

for note in notlar:
  if note >=80:
    yuksek_notlar.append(note)
  elif note <= 50:
    kalan += 1
  else:
    gecen += 1
  if note >= en_yuksek:
    en_yuksek = note
  toplam += note

ortalama = toplam / len(notlar)

print(f"Geçen    : {gecen} öğrenci")
print(f"Kalan    : {kalan} öğrenci")
print(f"En yüksek: {en_yuksek}")
print(f"Ortalama : {ortalama}")
````

## Soru 4 — Fonksiyonlar ve Sözlükler

Bir **kelime frekans sayacı** fonksiyonu yazın.

Fonksiyon bir cümle alıp her kelimenin kaç kez geçtiğini sözlük olarak döndürsün.

```python
cumle = "python öğren python sev python kullan ve python ile kod yaz"
```

**Beklenen çıktı:**
```
{'python': 4, 'öğren': 1, 'sev': 1, 'kullan': 1, 've': 1, 'ile': 1, 'kod': 1, 'yaz': 1}
En çok geçen kelime: 'python' (4 kez)
Toplam farklı kelime: 8
```

> 💡 İpucu: `split()` ile kelimelere böl. Sözlüğe eklerken `dict.get(kelime, 0)` kullan.

````python
cumle = "python öğren python sev python kullan ve python ile kod yaz"

# KODUNUZU BURAYA YAZIN
def kelime_say(cumle):
  cumle = cumle.split()
  frekans_sozluk = {}
  for kelime in cumle:
    frekans_sozluk[kelime] = cumle.count(kelime)

  print(frekans_sozluk)
  print(f"toplam farkli kelime:{len(frekans_sozluk.keys())}")
  print(f"En cok gecen kelime:{max(frekans_sozluk, key=frekans_sozluk.get)}")

kelime_say(cumle)

````

## Soru 5 — Hata Yönetimi (try / except)

Aşağıdaki `guvenli_hesap` fonksiyonunu tamamlayın.

Fonksiyon iki parametre alsın: `sayi` (string olarak gelecek) ve `bolen` (string olarak gelecek).

Olası hatalar:
- `sayi` veya `bolen` sayıya çevrilemezse → `"Hata: Geçersiz sayı"` yazdır
- `bolen` sıfırsa → `"Hata: Sıfıra bölünemez"` yazdır
- Hata yoksa → sonucu yazdır
- Her durumda → `"İşlem tamamlandı."` yazdır

```python
# Test senaryoları:
guvenli_hesap("10", "2")      # Sonuç: 5.0  →  İşlem tamamlandı.
guvenli_hesap("10", "0")      # Hata: Sıfıra bölünemez  →  İşlem tamamlandı.
guvenli_hesap("abc", "2")     # Hata: Geçersiz sayı  →  İşlem tamamlandı.
```

> 💡 İpucu: `try / except ValueError / except ZeroDivisionError / finally` kullanın.

````python
def guvenli_hesap(sayi, bolen):
    try:
      a = float(sayi)
      b = float(bolen)
      sonuc = a / b
    except ValueError:
      print("Hata: Geçersiz sayı")
    except ZeroDivisionError:
      print("Hata: Sıfıra bölünemez")
    else:
      print(f"Sonuç: {sonuc}")

# Test et:
guvenli_hesap("10", "2")
print("---")
guvenli_hesap("10", "0")
print("---")
guvenli_hesap("abc", "2")
````
