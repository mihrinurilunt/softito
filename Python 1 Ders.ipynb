```python
class Dikdortgen:
    def __init__(self, en, boy):
        self.en  = en
        self.boy = boy

    def alan(self):
        return self.en * self.boy

    def cevre(self):
        return 2 * (self.en + self.boy)

    def bilgi(self):
        print(f"Alan: {self.alan()}, Çevre: {self.cevre()}")

d = Dikdortgen(5, 3)
d.bilgi()
print(d.alan())    # 15
print(d.cevre())   # 16
```

```python
# Bunların her biri ayrı bir instance:
d1 = Dikdortgen(5, 3)    # 1. instance
d2 = Dikdortgen(10, 7)   # 2. instance
d3 = Dikdortgen(2, 2)    # 3. instance
print(d1.en)
print(d2.en)
print(d3.en)

d1.alan()
d2.alan()
d3.alan()
```

```python
class Hayvan:
    def __init__(self, isim, yas):
        self.isim = isim
        self.yas  = yas

    def ses_cikar(self):
        return "..."

    def tanitim(self):
        print(f"{self.isim} ({self.yas} yaş): {self.ses_cikar()}")


class Kopek(Hayvan):   # Hayvan'dan türüyor
    def __init__(self, isim, yas, irk):
        super().__init__(isim, yas)
        self.irk = irk

    def ses_cikar(self):   # override
        return "Hav hav!"


class Kedi(Hayvan):
    def ses_cikar(self):
        return "Miyav!"


k = Kopek("Karabaş", 3, "Kangal")
c = Kedi("Pamuk", 5)

k.tanitim()
c.tanitim()
```

```python
class BankaHesabi:
    def __init__(self, sahip, bakiye=0):
        self.sahip   = sahip
        self._bakiye = bakiye   # dahili

    @property
    def bakiye(self):
        return self._bakiye

    @bakiye.setter
    def bakiye(self, miktar):
        if miktar < 0:
            raise ValueError("Bakiye negatif olamaz!")
        self._bakiye = miktar

    def para_yatir(self, miktar):
        self._bakiye += miktar
        print(f"+{miktar} TL → Bakiye: {self._bakiye} TL")

    def para_cek(self, miktar):
        if miktar > self._bakiye:
            print("Yetersiz bakiye!")
        else:
            self._bakiye -= miktar
            print(f"-{miktar} TL → Bakiye: {self._bakiye} TL")

h = BankaHesabi("Ali", 1000)
h.para_yatir(500)
h.para_cek(200)
h.para_cek(2000)
```

```python
class Vektor:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):        # print() için
        return f"({self.x}, {self.y})"

    def __repr__(self):       # geliştirici çıktısı
        return f"Vektor({self.x}, {self.y})"

    def __add__(self, other):  # + operatörü
        return Vektor(self.x + other.x, self.y + other.y)

    def __len__(self):         # len() için (boyut sayısı)
        return 2

    def uzunluk(self):
        return (self.x**2 + self.y**2) ** 0.5

v1 = Vektor(3, 4)
v2 = Vektor(1, 2)

print(v1)           # __str__
print(v1 + v2)      # __add__
print(len(v1))      # __len__
print(v1.uzunluk()) # 5.0
```

```python
class Tarih:
    def __init__(self, gun, ay, yil):
        self.gun = gun
        self.ay  = ay
        self.yil = yil

    def __str__(self):
        # kullanıcı dostu — güzel görünsün
        return f"{self.gun} {self.ay} {self.yil}"

    def __repr__(self):
        # geliştirici dostu — nesneyi yeniden yaratabileyim
        return f"Tarih({self.gun}, {self.ay}, {self.yil})"


t = Tarih(15, 6, 2024)
```

```python
class Ogrenci:
    okul = "Atatürk Lisesi"   # class değişkeni (ortak)
    kayitli_sayi = 0

    def __init__(self, ad, not_):
        self.ad   = ad          # instance değişkeni
        self.not_ = not_
        Ogrenci.kayitli_sayi += 1

    def bilgi(self):
        print(f"{self.ad} | {self.okul} | Not: {self.not_}")


o1 = Ogrenci("Ayşe", 95)
o2 = Ogrenci("Mehmet", 78)

o1.bilgi()
o2.bilgi()
print("Toplam öğrenci:", Ogrenci.kayitli_sayi)

# Class değişkeni tüm örnekleri etkiler
Ogrenci.okul = "Cumhuriyet Lisesi"
o1.bilgi()   # okul değişti!
```
